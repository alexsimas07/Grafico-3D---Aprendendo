# Exemplo básico - gráfico em 3D wireframe plot

# Curso Cientista de dados - python 

```python
# importando as bibliotecas necessárias

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
```

```python
# Criando a figura e projeção em 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X, Y, Z = axes3d.get_test_data(0.10)
ax.plot_wireframe(X, Y, Z, rstride=15, cstride=15)
plt.show()
```

# Exemplo básico - gráfico Filled Contours

```python
# Importando as bibliotecas necessárias
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
# Criando a figura e projeção em 3D
fig = plt.figure()
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.15)
cset = ax.contourf(X, Y, Z, cmap = cm.coolwarm)
plt.show()
```

# Exemplo básico - Gráfico Plotting Contour (Utilizando extend3d)

```python
# Importando as bibliotecas necessárias
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
# Criando a figura e projeção em 3D
fig = plt.figure()
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)
cset = ax.contour(X, Y, Z, extend3d=True, cmap=cm.seismic)
plt.show()
```
